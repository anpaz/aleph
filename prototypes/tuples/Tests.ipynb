```qsharp
open Microsoft.Quantum.Diagnostics;

operation Apply_oracle(
    f: (Qubit[], Qubit[], Qubit[]) => Unit is Adj,
    registers: Range[],
    qubits: Qubit[],
    target: Qubit
    ): Unit 
is Adj {

    let x = qubits[registers[0]];
    let y = qubits[registers[1]];
    let z = qubits[registers[2]];
    
    use a = Qubit[2];
    
    within {
        f(x, y, a);
    } apply {
        AreEqual(a, z, target);
    }
}


operation AreEqual(x: Qubit[], y: Qubit[], answer: Qubit) : Unit
is Adj
{
    let l = Length(x);
    use a = Qubit[l];

    within {
        for i in 0..l-1 {
            CNOT(x[i], a[i]);
            CNOT(y[i], a[i]);
            X(a[i]);
        }
    } apply {
        Controlled X (a, answer);
    }
}

    
operation Add(x: Qubit[], y: Qubit[], answer: Qubit[]) : Unit
is Adj {
    //TODO: let l = Length(x);
    CNOT(x[0], answer[0]);
    CNOT(y[0], answer[0]);
    CCNOT(x[0], y[0], answer[1]);

    CNOT(x[1], answer[1]);
    CNOT(y[1], answer[1]);
}

operation main() : Unit {
    use x = Qubit[2];
    use y = Qubit[2];
    use z = Qubit[2];
    use target = Qubit();
    let r = [0..1,2..3,4..5];
    
    let all = x + y + z;
    
    ApplyToEach(H, all);

    Apply_oracle(Add, r, all, target);
    Message($"all: {all}");
    DumpRegister((), all + [target]);
    ResetAll(all + [target]);
}
```
